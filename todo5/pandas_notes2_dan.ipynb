### Pandas Notes part 2

```python
import pandas as pd
```

```python
df = pd.read_csv('mlb_standings_2016_katherine.csv')
print(df)
```

```python
import matplotlib.pyplot as plt
df.plot(kind='hist', stacked=True)
plt.show()
```

```python
df.plot(kind='scatter', x='Wins', y='Losses' ) # should be linear
plt.show()
```

```python
df[df['Division'] == 'Central'].plot(kind='bar', y='Wins', x = 'Team')
plt.show()
```

```python
df[df['League'] == 'National']['Wins'].plot(kind='pie', labels=df['Team'])
plt.show()
```

```python
# isn't too much numerical data, so just trying to group things and make more of the same plots
df['Total'] = df['Wins'] + df['Losses']

print(df['Total'])
```
