Data used comes from a csv file 'NHL 2016-17 Player Data' downloaded from http://www.hockeyabstract.com/testimonials/nhl2016-17playerdata.

Data for this csv was collected from a wide variety of analysts and web sites including NHL.com, Hockey Reference, Emmanuel Perry's Corsica Hockey, David Johnson and Hockey Analysis, Craig Tabita's Puck Base, Sporting Charts, Nick Abe's Xtra Hockey Stats, plus Dominik Luszczszyn's Game Score, and Thomas Crawshaw's injury data.

The file has 167 columns containing data on 888 players; every player that played atleast one game in the 2016-17 season.  

Columns of interest include index 0 'Born' (Date of birth), index 5 'Ht'(Height in inches), index 6 'Wt' (Weight in pounds), index 12 'Last Name', index 13 'First Name', index 16 'GP' (Games Played), index 17 'G' (Goals Scored), index 18 'A' (Number of Assists)and index 21 'PTS' (total points).

While NumPy has efecient functions and methods for dealing with arrays of the same data type, most data will come in files of mixed data types.  

Pandas is used to create objects called dataframes.  Dataframes are two demensional arrays where each column can be of a different data type.

Create a new dataframe of just relative data (np_rel_data)

```python
pd_rel_data = nhl.loc[:, ('Last Name', 'First Name', 'GP', 'G', 'A', 'PTS')]
```

```python
pd_rel_data.head()
```

Only interested in players that have played atleast 41 games in the season.

```python
pd_rel_data = pd_rel_data[pd_rel_data.GP >= 41]
```

```python
pd_rel_data.head()
```

Add a column 'PPG' (Points per Game) to the dataframe.  'PTS'/'GP'

```python
pd_rel_data = pd_rel_data.assign(PPG = pd_rel_data['PTS']/pd_rel_data['GP'])
```
