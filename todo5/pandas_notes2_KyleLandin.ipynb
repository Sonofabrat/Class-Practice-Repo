```python
filepath = '2010_Census_KyleLandin.csv'
census = pd.read_csv(filepath)
census.info()
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
males = census[["Total Males"]]
females = census[["Total Females"]]
med_age = census[["Median Age"]]
pop = census[["Total Population"]]
tot_house = census[["Total Households"]]
avg_house = census[["Average Household Size"]]
```

```python
%matplotlib inline
# Can specify what you want the plot points to look like. g^ will be green triangles, o will make dots, s will make squares, etc.
plt.plot(med_age, males, 'g^')
plt.title("Male Median Age")
plt.xlabel("Median Age")
plt.ylabel("Population")
```

```python
plt.plot(females,med_age,'bs')
plt.title("Female Median Age")
plt.xlabel("Population")
plt.ylabel("Medain Age")
```

```python
plt.plot(avg_house,tot_house,'ro')
plt.title("Average Household Size")
plt.xlabel("Household Size")
plt.ylabel("Total Households")
```
