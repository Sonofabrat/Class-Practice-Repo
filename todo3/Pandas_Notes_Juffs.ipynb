You can build DATA FRAMES in Pandas from lists, as follows. Note that the dictionary can summarized for 'dict' using the names of the lists and the values do not have to be retyped. Remember that booleans and floats do not have '...', whereas str MUST have '...'.

```python
import pandas as pd
# Build cars DataFrame
names = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr =  [True, False, False, False, True, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]
dict = { 'country':names, 'drives_right':dr, 'cars_per_cap':cpc }
cars = pd.DataFrame(dict)

# If you print the result of this code, you get numbered row labels. You can add them thus:
# Definition of row_labels
row_labels = ['US', 'AUS', 'JAP', 'IN', 'RU', 'MOR', 'EG']

# Specify row labels of cars
cars.index = row_labels

# Print cars again with the row labels
print(cars)
```
