## (6) Create another column: lyric length in word count
- All symbols have been removed from lyrics
- So, `.split()` works as a rudimentary tokenizer

```python
#songs.head()
len(songs.loc[0,'Lyrics'].split())

#YOU HAVE TO GET RID OF THE NaN!!!
#LOOK BELOW FOR MORE

#length of lyrics for first 50
# foo = songs[0:50]
# for x in foo['Lyrics']:
#     print(len(x.split()))

#list of token lengths
toklengths = [len(x.split()) for x in songs['Lyrics']]
#toklengths[:100]

songs['Length'] = pd.Series(toklengths)
songs.head()

#defining a function to split
def tokenize(txt):
    return txt.split()

#using the function to make a column of all the tokenized lyrics
songs['Tokens'] = songs['Lyrics'].map(tokenize)
songs.head()

#same thing as above but slightly different (the same thigns without defiing a function)
songs['Tokens2'] = songs['Lyrics'].map(lambda x: x.split())
songs.head()

#delete a column
del songs['Tokens2']
songs.head()

#calculate individual word lengths
#start with a function to calculate average word length
def get_avg_len(toks):
    lens = [len(t) for t in toks]
    if len(toks) == 0: return 0
    else: return sum(lens)/len(lens)

#function check
sent = 'Colorless green ideas sleep furiously'.split()
get_avg_len(sent)

#Adding average word length
songs['Avg Wrd Length'] = songs['Tokens'].map(get_avg_len)
songs.head()

#example of ttr
rose = 'rose is a rose is a rose is a rose.'.split()
rose
len(set(rose))

#defining the ttr function
def get_ttr(toks):
    if len(toks) == 0: return 0
    else: return len(set(toks))/len(toks)

get_ttr(rose)

#Adding ttr column to songs
songs["TTR"] = songs['Tokens'].map(get_ttr)
songs.head()
```
